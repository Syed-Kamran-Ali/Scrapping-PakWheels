## Importing Libraries

```python
import re

import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service

from webdriver_manager.chrome import ChromeDriverManager

import time
import re
```

## Opening Website

```python
ser = Service(r'C:\Chrome WebDriver\chromedriver.exe')
op = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=ser, options=op)

website = 'https://www.pakwheels.com'
driver.get(website)
driver.maximize_window()
```

## Clicking AND SEARCHING 

```python

search = driver.find_element(By.CLASS_NAME,"ui-autocomplete-input")
search.click()
time.sleep(3)
search.send_keys("Toyota Corolla")
search.send_keys(Keys.ENTER)
```

```python

title = []
city = []
model_year = []
mileage = []
fuel_type = []
engine_size = []
transmission = []
feature_lst = []
price = []
update = []
image=[]

for i in range(1,100+1):
    time.sleep(5)

    title_xpath = '/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[2]/div[1]/div/div/a/h3'
    city_xpath = '/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[2]/div[2]/div/ul[1]/li'
    year_xpath = "/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[2]/div[2]/div/ul[2]/li[1]"
    mileage_xpath = "/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[2]/div[2]/div/ul[2]/li[2]"
    fuel_type_xpath = "/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[2]/div[2]/div/ul[2]/li[3]"
    engine_size_xpath = "/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[2]/div[2]/div/ul[2]/li[4]"
    transmission_xpath = "/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[2]/div[2]/div/ul[2]/li[5]"
    featured_xpath = "/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[1]"
    price_xpath = '/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[2]/div[1]/div/div/div/div'
    update_xpath = '/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[2]/div[3]/div[1]'
    image_xpath = '/html/body/div[6]/section[2]/div/div[3]/div[2]/div/div[2]/div[2]/ul/li/div/div[1]/div[1]/div/ul/li/img'



    for i in driver.find_elements(By.XPATH,title_xpath):
        title.append(i.text)
    for j in driver.find_elements(By.XPATH,city_xpath):
        city.append(j.text)
    for k in driver.find_elements(By.XPATH,year_xpath):
        model_year.append(k.text)
    for l in driver.find_elements(By.XPATH,mileage_xpath):
        mileage.append(l.text)
    for m in driver.find_elements(By.XPATH,fuel_type_xpath):
        fuel_type.append(m.text)
    for n in driver.find_elements(By.XPATH,engine_size_xpath):
        engine_size.append(n.text)
    for o in driver.find_elements(By.XPATH,transmission_xpath):
        transmission.append(o.text)
    for p in driver.find_elements(By.XPATH,featured_xpath):
        try:
            feature = p.find_element(By.CLASS_NAME,"inner")
            feature_lst.append(feature.text)
        except:
            feature_lst.append("Not Featured")
    for q in driver.find_elements(By.XPATH,price_xpath):
        price.append(q.text)
    for r in driver.find_elements(By.XPATH,update_xpath):
        update.append(r.text)
    for s in driver.find_elements(By.XPATH,image_xpath):
        image.append(s.get_attribute("src"))

    driver.find_element(By.CLASS_NAME,"next_page").click()



```

```python
data = {"title" : title,
        "city" : city,
        "model_year" : model_year,
        "mileage" : mileage,
        "fuel_type" : fuel_type,
        "engine_size" : engine_size,
        'transmission' : transmission,
        "feature_lst" : feature_lst,
        "price" : price,
        "update" : update,
        "image": image}
```

```python
df = pd.DataFrame({ key:pd.Series(value) for key, value in data.items() })
```
